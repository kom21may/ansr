```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv("stock_prices.csv")

# Convert the date column to a datetime object
df["date"] = pd.to_datetime(df["date"])

# Calculate the rolling 30-day moving average
closing_price_ma = df["closing_price"].rolling(window=30).mean()

# Plot the closing price and the rolling 30-day moving average
plt.plot(df["date"], df["closing_price"], label="Closing Price")
plt.plot(df["date"], closing_price_ma, label="Rolling 30-Day MA")
plt.legend()
plt.show()
```
